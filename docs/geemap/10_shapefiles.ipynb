```python
# Initialize Earth Engine
import ee
try:
    ee.Initialize()
except Exception as e:
    ee.Authenticate()
    ee.Initialize(project="ee-zekuwany")
```

```python
Map = geemap.Map()
Map
```

```python
countries_shp = '../data/Oromia_Region.shp'
countries = geemap.shp_to_ee(countries_shp)
Map.addLayer(countries, {}, 'Countries')
```

```python
states_shp = '../data/Easrt_HZ.shp'
states = geemap.shp_to_ee(states_shp)
Map.addLayer(states, {}, 'US States')
```

```python
cities_shp = '../data/Gara_Mullata_Park.shp'
cities = geemap.shp_to_ee(cities_shp)
Map.addLayer(cities, {}, 'US Cities')
```

```python
geemap.ee_to_shp(countries, filename='../data/Easrt_HZ_new.shp')
```

```python
geemap.ee_export_vector(states, filename='../data/Gara_Mullata_Park.csv')
```

```python
geemap.ee_export_vector(states, filename='../data/Gara_Mullata_Park.kml')
```
